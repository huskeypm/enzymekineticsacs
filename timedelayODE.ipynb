## One component negative feedback model $ \frac{dy}{dt} = \frac{1}{1+(Y(t-d)/k)} - rY(t)$

```python
# MODEL, WITH UNKNOWN PARAMETERSmodel = lambda Y,t,k,d,r :  1/(1+(Y(t-d)/k)**2) - r*Y(t) # HISTORYg = lambda t:0    
```

### Impact of time-delay (e.g. diffusion between compartments) on oscillator frequency

```python
# SOLVINGtt = linspace(0,50,10000)yy4p5 = ddeint(model,g,tt,fargs=( 0.1 , 4.5 , 1. )) # K = 0.1, d = 5, r = 1yy5p0 = ddeint(model,g,tt,fargs=( 0.1 , 5 , 1. )) # K = 0.1, d = 5, r = 1yy5p5 = ddeint(model,g,tt,fargs=( 0.1 , 5.5 , 1. )) # K = 0.1, d = 5, r = 1 
```

Here we see that frequency is negatively correlated with delay (diffusion between compartments). Higher diffusion constants reduce delay, thus frequency is positively correlated with diffusion rate

```python
# PLOTTINGfig,ax=subplots(1)ax.plot(tt,yy4p5,lw=2,label="d=4.5 [s]")ax.plot(tt,yy5p0,lw=2,label="d=5.0 [s]")ax.plot(tt,yy5p5,lw=2,label="d=5.5 [s]")plt.ylabel('[Y] [M]')plt.xlabel('t [s]')plt.title("Time delay") plt.gcf().savefig("timedelayODE.png") legend(loc=0)
```

Can long delays still sustain oscillations? (yes) 

```python
ttl = linspace(0,500,10000)yy50p0 = ddeint(model,g,tt,fargs=( 0.1 , 50.0 , 1. )) # K = 0.1, d = 5, r = 1yy20p0 = ddeint(model,g,tt,fargs=( 0.1 , 20.0 , 1. )) # K = 0.1, d = 5, r = 1
```

```python
# PLOTTINGfig,ax=subplots(1)#ax.plot(tt,yy4p5,lw=2,label="d=4.5")ax.plot(tt,yy5p0,lw=2,label="d=5.0")ax.plot(ttl,yy20p0,lw=2,label="d=20.0")ax.plot(ttl,yy50p0,lw=2,label="d=50.0")legend(loc=0)plt.ylabel('[Y] [M]')plt.xlabel('t [s]')plt.title("kon") plt.gcf().savefig("konODE.png") 
```

### Impact of kon (using K=KD=koff/kon) on oscillator

```python
tt = linspace(0,50,10000)koff = 1.kon=20. ; K = koff/konprint Kyy5p0Kred = ddeint(model,g,tt,fargs=( K , 5 , 1. )) # K = 0.1, d = 5, r = 1kon=10. ; K = koff/konprint Kyy5p0 = ddeint(model,g,tt,fargs=( K , 5 , 1. )) # K = 0.1, d = 5, r = 1kon=5. ; K = koff/konprint Kyy5p0Kincr = ddeint(model,g,tt,fargs=( K , 5 , 1. )) # K = 0.1, d = 5, r = 1
```

Frequency is inversely correlated with kon. Here we assume that diffusion only affects kon (arrival of substrate to enzyme) but there are no other lags in system. This corresponds to the case where the system is well-mixed
