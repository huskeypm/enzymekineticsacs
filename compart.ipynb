```python
# Compart1: A<-->B, y0,y1# Compart2# src - apply periodic source term def twoCompmodel(t,y0,params,src=1,title=""):  def f(y,t,params):           # A<-->B     #s=0    s=1.    rxnTerms = s*dot(params.a,y)            # diffusion terms    A1 = y[0]; A2 = y[2];     B1 = y[1]; B2 = y[3]    DA=params.DA    DB=params.DB    volCompart1 = params.volCompart1    volCompart2 = params.volCompart2    saCompart1 = params.saCompart1    saCompart2 = params.saCompart2    L = 1 # see below for units     jA12 = -DA*(A1*saCompart1/volCompart1 - A2 *saCompart2/volCompart2)/L    jB12 = -DB*(B1*saCompart1/volCompart1 - B2 *saCompart2/volCompart2)/L    diffTerms = np.array([jA12,jB12,-jA12,-jB12])        # src terms     ky0 = params.c * src*(0.3*np.sin(params.f*t))    srcTerms = s*np.array([ky0,0,0,0])            dcdt = rxnTerms + srcTerms + diffTerms    return dcdt    #return [\    #        ky0+a[0,0]*y[0] + a[0,1]*y[1],\    #        a[1,0]*y[0] + a[1,1]*y[1],\  y = scipy.integrate.odeint(f,y0,t,args=(params,))  plot(t,y[:,0],"b-",label="A$_1$")  plot(t,y[:,1],"r-",label="B$_1$")  plot(t,y[:,2],"b--",label="A$_2$")  plot(t,y[:,3],"r--",label="B$_2$")      ylabel("Conc [$\mu$M]")  xlabel("t [s]")  legend(loc=0,ncol=2)      return y  
```
